Q1: What is Estimation Statistics? Explain point estimate and interval estimate.


Estimation statistics is a branch of statistics that involves estimating unknown population parameters based on sample data. It is used to make inferences or draw conclusions about a population by using information obtained from a sample.

Point Estimate:
A point estimate is a single value that is used to estimate an unknown population parameter. It is obtained by using a statistic, such as the sample mean or sample proportion, to estimate the corresponding population parameter, such as the population mean or population proportion. Point estimates provide a specific value that represents the best guess or approximation of the unknown parameter based on the available sample data.

For example, if we want to estimate the average height of all students in a school, we can collect a sample of heights from a subset of students and calculate the sample mean height. This sample mean would serve as the point estimate for the population mean height.

Interval Estimate:
An interval estimate, also known as a confidence interval, provides a range of values within which the unknown population parameter is likely to lie. It takes into account the variability in the sample data and provides a measure of uncertainty associated with the point estimate.

A confidence interval is typically expressed as a range of values with an associated confidence level. For example, a 95% confidence interval for the population mean height of students might be (165 cm, 175 cm). This means that we can be 95% confident that the true population mean height falls within this interval.

Interval estimates are useful because they provide a range of values rather than a single point, which gives a better sense of the possible range of values for the population parameter. The width of the interval reflects the level of uncertainty, with wider intervals indicating greater uncertainty and narrower intervals indicating more precise estimates.

```python
#Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation


import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Calculate the standard error of the mean
    standard_error = sample_std_dev / math.sqrt(sample_size)
    
    # Calculate the margin of error (assuming a 95% confidence level)
    margin_of_error = 1.96 * standard_error
    
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    # Return the estimated population mean and confidence interval
    return sample_mean, lower_bound, upper_bound
```


Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.


Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (HA), and collecting evidence from the sample to determine which hypothesis is supported by the data.

The null hypothesis typically represents the status quo or the absence of an effect, while the alternative hypothesis represents the presence of an effect or a hypothesis of interest. The goal of hypothesis testing is to assess the evidence against the null hypothesis and make a decision whether to accept or reject it in favor of the alternative hypothesis.

Hypothesis testing is used for several reasons:

Testing Research Questions: Hypothesis testing allows researchers to test specific research questions or hypotheses. It helps in evaluating theories, investigating relationships between variables, and determining the validity of assumptions.

Decision Making: Hypothesis testing provides a systematic framework for decision making based on evidence from the sample. It helps in making informed decisions in various fields, such as medicine, business, social sciences, and more.

Scientific Rigor: Hypothesis testing adds rigor to the scientific process by providing a structured approach to assess the strength of evidence and draw conclusions. It helps in avoiding subjective interpretations and biases in data analysis.

Generalization: By testing hypotheses on sample data, hypothesis testing allows for generalizing the findings to a larger population. It helps in making inferences about the population based on the evidence obtained from the sample.

Q4. Create a hypothesis that states whether the average weight of male college students is greater than
the average weight of female college students.


Hypothesis:
The average weight of male college students is greater than the average weight of female college students.

Null Hypothesis (H0):
The average weight of male college students is equal to or less than the average weight of female college students.

Alternative Hypothesis (HA):
The average weight of male college students is greater than the average weight of female college students.

Symbolically, the hypotheses can be represented as:

H0: μ_male <= μ_female
HA: μ_male > μ_female

Where:
H0 represents the null hypothesis,
HA represents the alternative hypothesis,
μ_male represents the population mean weight of male college students, and
μ_female represents the population mean weight of female college students.

To test this hypothesis, a sample of weights from male and female college students would need to be collected. The data would then be analyzed using appropriate statistical methods, such as a t-test or a z-test, to determine if there is sufficient evidence to reject the null hypothesis in favor of the alternative hypothesis.

```python
#. Write a Python script to conduct a hypothesis test on the difference between two population means given a sample from each population


import numpy as np
from scipy.stats import t

def hypothesis_test(sample1, sample2, alpha):
    # Compute sample statistics
    n1 = len(sample1)
    n2 = len(sample2)
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    
    # Compute the test statistic (assuming unequal variances)
    dof = n1 + n2 - 2
    pooled_std = np.sqrt((std1**2 / n1) + (std2**2 / n2))
    t_value = (mean1 - mean2) / pooled_std
    
    # Compute the critical value and p-value
    crit_value = t.ppf(1 - alpha, dof)
    p_value = 2 * (1 - t.cdf(np.abs(t_value), dof))
    
    # Perform the hypothesis test
    if t_value > crit_value:
        print("Reject the null hypothesis. There is sufficient evidence to support the alternative hypothesis.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")
    
    # Print the test statistics and p-value
    print("Test statistic:", t_value)
    print("Critical value:", crit_value)
    print("P-value:", p_value)
```

State the Hypotheses: Clearly define the null hypothesis (H0) and the alternative hypothesis (HA) based on the research question or objective.

Calculate the p-value: Calculate the probability (p-value) of observing the test statistic or more extreme values, assuming the null hypothesis is true. The p-value represents the strength of evidence against the null hypothesis.

Make a Decision: Compare the p-value with the significance level (alpha). If the p-value is less than alpha, reject the null hypothesis in favor of the alternative hypothesis. If the p-value is greater than or equal to alpha, fail to reject the null hypothesis.

Q8. Define p-value and explain its significance in hypothesis testing.
In hypothesis testing, the p-value is a measure of the strength of evidence against the null hypothesis. It quantifies the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis is true.

The significance of the p-value in hypothesis testing can be understood as follows:

If the p-value is small (typically below the chosen significance level, often denoted as alpha), it suggests that the observed data are unlikely to have occurred by chance alone, assuming the null hypothesis is true. In this case, the evidence against the null hypothesis is considered strong, and we may choose to reject the null hypothesis in favor of the alternative hypothesis.

Conversely, if the p-value is large (greater than or equal to the chosen significance level), it suggests that the observed data are reasonably likely to have occurred by chance, assuming the null hypothesis is true. In this case, the evidence against the null hypothesis is not strong enough, and we fail to reject the null hypothesis.

It's important to note that the p-value does not provide direct information about the truth or falsity of the null hypothesis or the magnitude of any effect. It only informs us about the strength of evidence against the null hypothesis based on the observed data.

The significance of the p-value in hypothesis testing lies in its ability to help researchers make informed decisions. By comparing the p-value to the chosen significance level, researchers can determine whether the observed data provide sufficient evidence to reject the null hypothesis and support the alternative hypothesis or not.

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
 # random samples of equal size and a null hypothesis that the population means are equal.

import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples
sample1 = np.random.normal(loc=5, scale=2, size=100)
sample2 = np.random.normal(loc=7, scale=2, size=100)

# Perform the two-sample t-test
t_statistic, p_value = ttest_ind(sample1, sample2)

# Print the results
print("Two-Sample T-Test")
print("Sample 1 mean:", np.mean(sample1))
print("Sample 2 mean:", np.mean(sample2))
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

Q11: What is Student’s t distribution? When to use the t-Distribution.

The Student's t-distribution, often simply referred to as the t-distribution, is a probability distribution that is used in statistical inference for hypothesis testing and constructing confidence intervals when the sample size is small and the population standard deviation is unknown. It is similar to the standard normal distribution (Z-distribution) but has heavier tails.

The t-distribution is characterized by its degrees of freedom (df), which determine the shape of the distribution. As the degrees of freedom increase, the t-distribution approaches the shape of the standard normal distribution. The t-distribution has a mean of 0 and a variance of df/(df-2) when df > 2.

The t-distribution is used when:

The population standard deviation is unknown: When the population standard deviation is not known, it needs to be estimated from the sample. The t-distribution takes into account the additional uncertainty introduced by estimating the population standard deviation.

Small sample sizes: The t-distribution is particularly useful when dealing with small sample sizes (typically less than 30) where the sample size is insufficient to rely on the central limit theorem for approximation.

Normality assumption: The t-distribution assumes that the underlying population follows a normal distribution. While the assumption is not strictly required, it is often satisfied in practice, especially for large sample sizes.

Q12: What is t-statistic? State the formula for t-statistic.


The t-statistic, also known as the t-value, is a measure used in hypothesis testing to assess the difference between a sample mean and a population mean when the population standard deviation is unknown. It quantifies the difference between the observed sample mean and the hypothesized population mean in terms of standard errors.

The formula for the t-statistic is as follows:

t = (sample mean - hypothesized population mean) / (standard error)

Where:

"sample mean" is the mean value calculated from the sample data,
"hypothesized population mean" is the value assumed for the population mean under the null hypothesis,
"standard error" represents the standard deviation of the sample mean and is calculated as the sample standard deviation divided by the square root of the sample size.
The t-statistic measures how many standard errors the sample mean is away from the hypothesized population mean. It allows us to determine the significance of the observed difference and make conclusions about the null hypothesis.

In hypothesis testing, the t-statistic is compared to a critical value from the t-distribution with a specific degrees of freedom. If the absolute value of the t-statistic is larger than the critical value, it indicates that the observed difference is statistically significant, providing evidence against the null hypothesis.

By calculating the t-statistic and comparing it to critical values, we can make informed decisions about whether to reject or fail to reject the null hypothesis based on the observed sample data.

Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.


To estimate the population mean revenue with a 95% confidence interval, we can use the formula for a confidence interval based on the t-distribution. Here are the steps to calculate it:

Step 1: Identify the necessary values:

Sample mean (x̄) = $500
Sample standard deviation (s) = $50
Sample size (n) = 50
Confidence level = 95% (which corresponds to a significance level of α = 0.05)
Step 2: Calculate the standard error:

Standard error (SE) = s / √n

SE = 50 / √50 ≈ 7.07

Step 3: Determine the critical value:

Since we have a sample size of 50, the degrees of freedom will be (n - 1) = 49.
Look up the critical value associated with a 95% confidence level and 49 degrees of freedom in the t-distribution table. Let's assume it is approximately 2.009 (for a two-tailed test).
Step 4: Calculate the margin of error:

Margin of error (ME) = critical value * standard error

ME = 2.009 * 7.07 ≈ 14.20

Step 5: Calculate the confidence interval:

Lower bound = sample mean - margin of error

Upper bound = sample mean + margin of error

Lower bound = 500 - 14.20 ≈ 485.80
Upper bound = 500 + 14.20 ≈ 514.20

The 95% confidence interval estimate for the population mean revenue is approximately $485.80 to $514.20. This means we are 95% confident that the true population mean revenue falls within this range based on the given sample data.

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.


To test the hypothesis that the new drug decreases blood pressure by 10 mmHg, we can conduct a one-sample t-test. Here are the steps to perform the hypothesis test:

Step 1: Set up the hypotheses:

Null hypothesis (H₀): The new drug does not decrease blood pressure by 10 mmHg. μ = 0
Alternative hypothesis (H₁): The new drug decreases blood pressure by 10 mmHg. μ ≠ 0
Step 2: Define the significance level:

Significance level (α) = 0.05
Step 3: Calculate the test statistic:

Sample mean (x̄) = 8 mmHg
Hypothesized population mean (μ₀) = 10 mmHg
Sample standard deviation (s) = 3 mmHg
Sample size (n) = 100
The test statistic (t) is calculated using the formula:
t = (x̄ - μ₀) / (s / √n)

Plugging in the values:
t = (8 - 10) / (3 / √100)
t = -2 / 0.3
t ≈ -6.67

Step 4: Determine the critical value:

Since this is a two-tailed test with a significance level of 0.05, we divide the significance level by 2 to get 0.025.
Look up the critical value associated with a significance level of 0.025 and degrees of freedom (n - 1 = 99) in the t-distribution table. Let's assume it is approximately -1.984.
Step 5: Make a decision:

If the absolute value of the test statistic is greater than the absolute value of the critical value, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.
Since |-6.67| > |-1.984|, we reject the null hypothesis.

Step 6: Draw a conclusion:

Based on the sample data, there is sufficient evidence to conclude that the new drug decreases blood pressure by a mean difference of 10 mmHg, with a significance level of 0.05.

Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.


To test the hypothesis that the population means for the two groups are equal, we can conduct a two-sample t-test. Here are the steps to perform the hypothesis test:

Step 1: Set up the hypotheses:

Null hypothesis (H₀): The population means for the two groups are equal. μ₁ = μ₂
Alternative hypothesis (H₁): The population means for the two groups are not equal. μ₁ ≠ μ₂
Step 2: Define the significance level:

Significance level (α) = 0.01
Step 3: Calculate the test statistic:

Group 1 sample mean (x̄₁) = 80

Group 1 sample standard deviation (s₁) = 10

Group 1 sample size (n₁) = 30

Group 2 sample mean (x̄₂) = 75

Group 2 sample standard deviation (s₂) = 8

Group 2 sample size (n₂) = 40

The test statistic (t) is calculated using the formula:
t = (x̄₁ - x̄₂) / √((s₁²/n₁) + (s₂²/n₂))

Plugging in the values:
t = (80 - 75) / √((10²/30) + (8²/40))
t = 5 / √((100/30) + (64/40))
t = 5 / √(3.33 + 1.6)
t ≈ 5 / √4.93
t ≈ 5 / 2.22
t ≈ 2.25

Step 4: Determine the critical value:

Since this is a two-tailed test with a significance level of 0.01, we need to divide the significance level by 2 to find each tail's critical value. Let's assume the critical value is approximately ±2.614.
Step 5: Make a decision:

If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.
Since |2.25| < 2.614, we fail to reject the null hypothesis.

Step 6: Draw a conclusion:

Based on the sample data, there is insufficient evidence to conclude that the population means for the two groups are significantly different, with a significance level of 0.01.

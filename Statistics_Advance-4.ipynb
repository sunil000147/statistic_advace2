Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.


The t-test and the z-test are both statistical tests used to make inferences about population parameters based on sample data. The key difference between them lies in the assumptions made about the population and the available information.

A t-test is used when the population standard deviation is unknown and must be estimated from the sample. It is suitable for small sample sizes (typically less than 30) and follows a t-distribution. The t-test is commonly used to compare means between two groups or to determine if a sample mean differs significantly from a known population mean.

Example scenario for using a t-test:
Let's say a researcher wants to investigate whether a new teaching method improves student performance. They randomly select two groups of students, one taught using the traditional method and the other taught using the new method. After collecting test scores from both groups, the researcher can perform a t-test to determine if there is a significant difference in the mean scores between the two groups.

On the other hand, a z-test is used when the population standard deviation is known or when the sample size is large (typically greater than 30), allowing for an approximation of the normal distribution. The z-test is commonly employed when comparing sample proportions or testing hypotheses about a population mean.

Example scenario for using a z-test:
Suppose a company claims that their product has a defect rate of less than 2%. To verify this claim, a quality control manager takes a large random sample of products and inspects them for defects. By conducting a z-test on the sample proportion of defects, the manager can determine if there is sufficient evidence to support or reject the company's claim.

In summary, the choice between a t-test and a z-test depends on the sample size and the availability of population standard deviation information. If the sample size is small or the population standard deviation is unknown, a t-test is appropriate. If the sample size is large and the population standard deviation is known or can be reasonably estimated, a z-test can be used.

Q2: Differentiate between one-tailed and two-tailed tests. short 


A one-tailed test is used when the alternative hypothesis specifies the direction of the effect, while a two-tailed test is used when the alternative hypothesis does not specify the direction.

In a one-tailed test, statistical significance is assessed by examining whether the observed data significantly deviate from the null hypothesis in only one direction (positive or negative). It focuses on detecting an effect in a specific direction.

In a two-tailed test, statistical significance is assessed by examining whether the observed data significantly deviate from the null hypothesis in either direction. It considers the possibility of an effect in both directions.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.


In hypothesis testing, Type 1 and Type 2 errors are two types of mistakes that can occur when making decisions about the null and alternative hypotheses.

Type 1 error, also known as a false positive, occurs when the null hypothesis is rejected even though it is true. In other words, it is the incorrect rejection of a true null hypothesis. It represents the situation where the test results lead to concluding that there is an effect or relationship when, in reality, there is none.

Example scenario of a Type 1 error:
Let's say a drug company is testing a new medication's effectiveness for reducing blood pressure. The null hypothesis is that the drug has no effect on blood pressure. However, due to a statistical fluctuation or other factors, the study incorrectly detects a significant decrease in blood pressure and rejects the null hypothesis. This would be a Type 1 error since the drug actually has no effect on blood pressure, but the test mistakenly concludes otherwise.

Type 2 error, also known as a false negative, occurs when the null hypothesis is not rejected even though it is false. In other words, it is the failure to reject a false null hypothesis. It represents the situation where the test fails to detect an effect or relationship when, in reality, there is one.

Example scenario of a Type 2 error:
Continuing with the previous example, suppose the new medication actually does reduce blood pressure, but the study fails to detect a significant decrease and does not reject the null hypothesis. This would be a Type 2 error since the drug does have an effect on blood pressure, but the test fails to conclude so.

Type 1 and Type 2 errors are inversely related, meaning that reducing one type of error increases the probability of the other. Researchers often aim to balance the two types of errors by choosing an appropriate significance level (alpha) and sample size to minimize both types of errors.

The theorem is stated as follows:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

P(A|B) is the probability of event A given event B has occurred (the posterior probability).
P(B|A) is the probability of event B given event A has occurred (the likelihood).
P(A) is the probability of event A occurring without any additional information (the prior probability).
P(B) is the probability of event B occurring without any additional information (the marginal likelihood).

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example. words 100


A confidence interval is a statistical range that provides an estimated range of values within which a population parameter, such as a mean or proportion, is likely to lie. It measures the uncertainty associated with sample estimates.

To calculate a confidence interval, you typically follow these steps:

Collect a representative sample from the population of interest.
Compute the sample statistic (e.g., mean, proportion) based on the sample data.
Determine the desired confidence level (e.g., 95%, 99%) that reflects your level of confidence in capturing the true population parameter.
Identify the appropriate critical value from the corresponding probability distribution (e.g., z-value for a normal distribution, t-value for a t-distribution), based on the desired confidence level and sample size.
Calculate the standard error, which quantifies the variability of the sample statistic.
Use the sample statistic, critical value, and standard error to calculate the lower and upper bounds of the confidence interval.
Example: Let's say you want to estimate the average height of adult males in a city. You collect a random sample of 100 males and find the mean height to be 175 cm, with a standard deviation of 5 cm.

To calculate a 95% confidence interval for the population mean height:

Sample mean: 175 cm
Sample size: 100
Desired confidence level: 95%
Critical value: For a 95% confidence level, the critical value for a normal distribution is approximately 1.96.
Standard error: Compute the standard error by dividing the sample standard deviation by the square root of the sample size. In this case, it would be 5 cm / sqrt(100) = 0.5 cm.
Confidence interval calculation: Multiply the standard error by the critical value and add/subtract the result from the sample mean. The lower bound would be 175 cm - (1.96 * 0.5 cm), and the upper bound would be 175 cm + (1.96 * 0.5 cm).
Therefore, the 95% confidence interval for the average height of adult males in the city would be (174.02 cm, 175.98 cm), indicating that we are 95% confident that the true population mean height falls within this interval.

Finally, we can calculate P(A|B) using Bayes' Theorem:
P(A|B) = (P(B|A) * P(A)) / P(B) = (0.05 * 0.5) / 0.075 = 0.333 (approximately 33.3%)
